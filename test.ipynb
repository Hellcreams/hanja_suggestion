```python
from unicodedata import normalize
```

```python
# 한국용 한자 -> chr(19968 ~ 40959)
f = open("./original/특별한한자.txt", encoding="utf-8")
hanjas = f.read().replace("\n", " ")
hanja_list = hanjas.split()
a = 0
sus_list = []
for h in hanja_list:
    if not (19968 <= ord(normalize("NFKC", h)) <= 40959):
        sus_list.append(h)
print(sus_list)
```
